# **2. Remoção das duplicatas**

```python
df['EMPRESA'] = df.index.str[:4]
df = df.drop_duplicates(subset=['EMPRESA'])
```

# **3. Fazendo nosso screening**

```python
# Filtrar empresas com patrimonio liquido positivo e outro filtro sobre dívida bruta / pl
```

```python
filtrado = df[(df.patrliq >0)&(df.divbpatr < 0.5)]
```

```python
divida_bruta_patr = filtrado.sort_values('divbpatr',ascending=True).head(50)
```

```python
divida_bruta_patr.head(5)
```

```python
# Verificação de empresas que ainda são negociadas
from datetime import datetime, timedelta

def is_traded_recently(symbol):
    try:
        # Get historical data for the stock
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period='1mo')

        # Get the last trading day
        last_trading_day = historical_data.index[-1].date()

        # Check if the last trading day is within the window of days
        today = datetime.now().date()
        window_start = today - timedelta(days=5)

        return window_start <= last_trading_day <= today
    except Exception:
        # If there is an exception, consider it as delisted
        return False
```

```python
divida_bruta_patr['ação'] = divida_bruta_patr.index + '.SA'
divida_bruta_patr['negociada'] = divida_bruta_patr['ação'].apply(is_traded_recently)
```

```python
empresas_filtradas = divida_bruta_patr[divida_bruta_patr['negociada'] == True]
```

```python
empresas_filtradas = empresas_filtradas.sort_values('roe',ascending = False).head(10)
```
